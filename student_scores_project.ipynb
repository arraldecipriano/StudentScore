## Cleaning

```python
# Getting a sum of all null values
df.isnull().sum()
```

```python
# Dropping unnamed colum
df = df.drop('Unnamed: 0', axis = 1)
print(df.head())
```

```python
# Changing WklyStudyHours column
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace('05-Oct', '5-10') # The error was that Excel perceived 5-10 as fifth of October
df.head()
```

## Analyzing and Visualizing

```python
# Plotting the gender distribution
plt.figure(figsize = (5, 5)) # Setting the size of the plot
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0])
plt.show() # Showing the plot

# The number of females students is greater that the male
```

```python
# Grouping the data by ParentEduc and aggregating the mean of MathScore, ReadingScore and WritingScore
gb = df.groupby('ParentEduc').agg({'MathScore' : 'mean',
                                  'ReadingScore' : 'mean',
                                  'WritingScore' : 'mean'})
print(gb)
```

```python
# Plotting the results above
plt.figure(figsize = (10, 6))
sns.heatmap(gb, annot = True) # annot = True to show the values on each square
plt.title("Relationship between parent's education and children's scores")
plt.show()

# With this plot I'm comparing the correlation of parents education with the children score, we conclude that the parent's education does have a heavy impact on the children's scores. The average difference is of 11
```

```python
# Grouping the data by ParentMaritalStatus and aggregating the mean of MathScore, ReadingScore and WritingScore
gb1 = df.groupby('ParentMaritalStatus').agg({'MathScore' : 'mean',
                                  'ReadingScore' : 'mean',
                                  'WritingScore' : 'mean'})
print(gb1)
```

```python
# Plotting the results above
plt.figure(figsize = (10, 6))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parent's marital status and children's scores")
plt.show()

# With this plot I'm comparing the correlation of parents marital status with the children score, we conclude that the parent marital status does not heavily affect on the children's scores. The average difference is of 1
```

```python
# Distribution of EthnicGroup
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()  # Subseting the df for each EthnicGroup
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

# Creating a pie chart to visualize the results
mlist = [groupA['EthnicGroup'],
        groupB['EthnicGroup'],
        groupC['EthnicGroup'],
        groupD['EthnicGroup'],
        groupE['EthnicGroup']] # Creating a list of the .loc results specifying the column we need
print(mlist)
labels = ['group A', 'group B', 'group C', 'group D', 'group E']
plt.pie(mlist, labels = labels, autopct = '%1.2f%%') # autopct = '%1.2f%%' to show the percentage of each group
plt.title('Distribution of Ethnic Groups')
plt.show()
```
